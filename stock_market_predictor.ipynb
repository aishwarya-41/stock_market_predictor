```python
import yfinance as yf

ticker = 'BIOCON.NS'

data = yf.download(ticker,period='5y',interval='1d')

data.head()
```

2. Lag features

```python
n_lags = 5

for i in range(1,n_lags+1):
    data[f'Lag_{i}_Close'] = data['Close'].shift(i)
    
data['Target_Next_Close'] = data['Close'].shift(-1)

data.head()
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data.isnull().sum()
```

```python
for i in range(1,n_lags+1):
    data[f'Lag_{i}_High'] = data['High'].shift(i)
    
data['Target_Next_High'] = data['High'].shift(-1)

data.head()
```

```python
for i in range(1,n_lags+1):
    data[f'Lag_{i}_Low'] = data['Low'].shift(i)
    
data['Target_Next_Low'] = data['Low'].shift(-1)

data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data
```

3. Adding Derived Indicators

```python
# Moving Avg
data['Simple_Moving_Avg_5'] = data['Close'].rolling(window=5).mean() # Average closing price of past 5 days

data['Simple_Moving_Avg_10'] = data['Close'].rolling(window=10).mean() 
```

```python
# Rolling Volatility
data['Volatality_5'] = data['Close'].rolling(window=5).std() # Movement of closing prices for the past 5 days - variance
```

```python
data['Daily_return'] = data['Close'].pct_change() # % change from yest = (today's close- yest)/yest close
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data
```

4. Target variables
- Target Close has already been done
- Target Direction 1 for Up and 0 for Down

```python
data = data.reset_index(drop=True)
```

```python
data['Target_Direction'] = (data['Target_Next_Close']>data['Close']).astype(int)
```

```python
data['Target_Direction'] = (
    data['Target_Next_Close'].to_numpy() > data['Close'].to_numpy()).astype(int)
```

```python
print(data['Target_Next_Close'].shape)
print(data['Close'].shape)
```

```python
data['Target_Direction'] = (
    (data['Target_Next_Close'].values.ravel() > data['Close'].values.ravel())
).astype(int)
```

```python
data
```

```python
n_rows = len(data)

split_point = int(0.8*n_rows)

train_data = data.iloc[:split_point]
test_data = data.iloc[split_point:]
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# ✅ 1. Prepare data
# Drop 'target_direction' column
data_cleaned = data.drop(columns=['Target_Direction'])

# Define feature and target columns
target_col = 'Target_Next_Close'  # Replace with actual target column name
feature_cols = [col for col in data_cleaned.columns if col != target_col]

# Convert to numpy arrays
data_values = data_cleaned[feature_cols].to_numpy()
target_values = data_cleaned[target_col].to_numpy()

# ✅ 2. Create sequences
look_back = 60

X = []
y = []

for i in range(look_back, len(data_values)):
    X.append(data_values[i - look_back:i])
    y.append(target_values[i])

X = np.array(X)
y = np.array(y)

print("X shape:", X.shape)
print("y shape:", y.shape)

# ✅ 3. Split train/test
train_size = int(len(X) * 0.8)

X_train = X[:train_size]
y_train = y[:train_size]

X_test = X[train_size:]
y_test = y[train_size:]

# ✅ 4. Scale features (fit only on training data)
n_timesteps = X_train.shape[1]
n_features = X_train.shape[2]

scaler = StandardScaler()

# Reshape for scaler
X_train_reshaped = X_train.reshape(-1, n_features)
X_train_scaled = scaler.fit_transform(X_train_reshaped)
X_train_scaled = X_train_scaled.reshape(-1, n_timesteps, n_features)

X_test_reshaped = X_test.reshape(-1, n_features)
X_test_scaled = scaler.transform(X_test_reshaped)
X_test_scaled = X_test_scaled.reshape(-1, n_timesteps, n_features)

print("X_train_scaled shape:", X_train_scaled.shape)
print("X_test_scaled shape:", X_test_scaled.shape)

# ✅ 5. Build LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(n_timesteps, n_features)))
model.add(Dropout(0.2))
model.add(LSTM(50))
model.add(Dropout(0.2))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')

# ✅ 6. Train model
history = model.fit(
    X_train_scaled, y_train,
    epochs=20,
    batch_size=32,
    validation_data=(X_test_scaled, y_test)
)

# ✅ 7. Evaluate
y_pred = model.predict(X_test_scaled).flatten()

r2 = r2_score(y_test, y_pred)
print("R^2 score:", r2)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import numpy as np

# Assuming you already have X and y from your sequence creation
print("Original y stats:", y.min(), y.max(), y.mean(), y.std())

# Scale X
X_flat = X.reshape(-1, X.shape[2])
scaler_X = StandardScaler()
X_flat_scaled = scaler_X.fit_transform(X_flat)
X_scaled = X_flat_scaled.reshape(X.shape)

# Scale y
scaler_y = StandardScaler()
y_scaled = scaler_y.fit_transform(y.reshape(-1,1)).flatten()

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_scaled, test_size=0.2, shuffle=False
)

# Model
model = Sequential()
model.add(LSTM(32, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(16))
model.add(Dense(1))

model.compile(optimizer="rmsprop", loss="mse")

history = model.fit(
    X_train, y_train, 
    validation_data=(X_test, y_test),
    epochs=50,
    batch_size=16,
    verbose=1
)

# Predict
y_pred_scaled = model.predict(X_test).flatten()
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1,1)).flatten()
y_true = scaler_y.inverse_transform(y_test.reshape(-1,1)).flatten()

# R2
r2 = r2_score(y_true, y_pred)
print("R2 Score:", r2)
```
